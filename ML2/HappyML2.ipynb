```python
HappinessCSV = pd.read_csv('YR2015_2019FinalML3.csv')
HappinessCSV
```

```python
HappinessCSV.drop_duplicates(subset="Country", keep="first", inplace=True)
#HappinessCSV.describe()
Happiness1=HappinessCSV
Happiness1
```

```python
Happiness1.loc[:,'Happiness2019str'] = Happiness1['Happiness2019']
Happiness1
```

```python
Happiness1['Happiness2020'] = " "
Happiness1
```

```python
#Happiness.replace(to_replace=r'^ba.$', value='new', regex=True)
#Happiness.replace('Happy','1')
#Happiness=Happiness.replace(to_replace="Happy",value="1")
#Happiness=Happiness.replace(to_replace="NotHappy",value="0")
#Happiness.Happiness2015[Happiness.Happiness2015 == 'Happy'] = 1
#Happiness.Happiness2015[Happiness.Happiness2015 == 'NotHappy'] = 0
#Happiness
```

```python
Happiness2=Happiness1.replace(to_replace="Happy",value="1")
Happiness=Happiness2.replace(to_replace="NotHappy",value="0")
Happiness
```

```python
Happiness.Happiness2019str[Happiness.Happiness2019str == '0'] = "NotHappy"
Happiness.Happiness2019str[Happiness.Happiness2019str == '1'] = "Happy" 
Happiness
```

```python
X = Happiness.drop(["Country","Happiness2020", "Happiness2019str"], axis=1)
y = Happiness["Happiness2019str"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
#Create a Deep Learning Model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Create model and add layers
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=40))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
#Quantify our Trained Model
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:150])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```
